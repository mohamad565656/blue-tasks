```python
def tokenization(names):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(names)
    return tokenizer
```

```python
fa_tokenizer = tokenization(aaa[:,0])
fa_vocab_size = len(fa_tokenizer.word_index)+1

fa_length = 2
print('persian name size: %d' % fa_vocab_size)
```

```python
en_tokenizer = tokenization(aaa[:,1])
en_vocab_size = len(en_tokenizer.word_index)+1

en_length = 2
print('english name size: %d' % en_vocab_size)
```

```python
def encode_sequences(tokenizer, length, names):
    seq = tokenizer.texts_to_sequences(names)
    seq = pad_sequences(seq, maxlen=length, padding = 'post')
    return seq
```

```python
train , test = train_test_split(aaa, test_size = 0.2, random_state = 8)
```

```python
x_train = encode_sequences(fa_tokenizer, fa_length, train[:,0])
y_train = encode_sequences(en_tokenizer, en_length, train[:,1])
```

```python
x_test = encode_sequences(fa_tokenizer, fa_length, test[:,0])
y_test = encode_sequences(en_tokenizer, en_length, test[:,1])
```

```python
def define_model(input_, output_, in_timesteps, out_timesteps, units):
    model = Sequential()
    model.add(Embedding(input_, units, input_length = in_timesteps, mask_zero= True))
    model.add(LSTM(units))
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    model.add(RepeatVector(out_timesteps))
    model.add(LSTM(units, return_sequences=True))
    model.add(Dense(output_, activation='softmax'))
    return model
```

```python
model = define_model(fa_vocab_size, en_vocab_size, fa_length, en_length, 512 )
model.compile(optimizer=Adam(learning_rate= 0.001) , loss = 'sparse_categorical_crossentropy')
```

```python
history = model.fit(x_train, y_train.reshape(y_train.shape[0],y_train.shape[1],1),
                   epochs = 70 , batch_size = 512, validation_split = 0.2)
```

```python
predict_x=model.predict(x_test.reshape((x_test.shape[0], x_test.shape[1]))) 
classes_x=np.argmax(predict_x,axis=1)
```

```python
def get_word(n, tokenizer):
    for word, index in tokenizer.word_index.items():
        if index == n:
            return word
    return None
```

```python
pred_text = []
for i in classes_x:
    temp = []
    for j in range(len(i)):
        t = get_word(i[j], en_tokenizer)
        if j > 0:
            if (t== get_word(i[j-1], en_tokenizer)) or (t == None):
                temp.append('')
            else:
                temp.append(t)
    pred_text.append(''.join(temp))
```
